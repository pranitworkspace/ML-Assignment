1. What are the key tasks that machine learning entails? What does data pre-processing imply?

The tasks which play import role in machine learning include data collection, data pre-processing, feature engineering, model selection, model training, model evaluation, and prediction.
Data pre-processing can be the steps taken to clean, transform, and prepare the data for machine learning. It involves handling missing values, dealing with outliers, standardizing or normalizing data, encoding categorical variables, and splitting the data into training and testing sets.

2. Describe quantitative and qualitative data in depth. Make a distinction between the two.

Quantitative data: Quantitative data represents numerical values that can be measured or counted. It includes variables such as age, height, weight, temperature, and income. This data can be further classified as discrete (countable) or continuous (measurable on a scale).

Qualitative data: Qualitative data represents non-numerical information and is descriptive in nature. It includes variables such as color, gender, occupation, and opinions. This data is often categorical and can be further divided into nominal (no inherent order or ranking) and ordinal (ordered or ranked categories).

6. How would the learning activity be affected if certain variables have missing values? Having said that, what can be done about it?

If certain variables have missing values, it can affect the learning activity by introducing bias or reducing the accuracy of the models. Missing values can lead to incomplete or incorrect patterns, and some machine learning algorithms may not handle missing data directly.

To handle missing values:

Remove instances with missing values: If the missing values are small in number, removing the instances with missing values can be an option.
Fill missing values with statistical measures: Missing values can be imputed with measures such as mean, median, mode, or regression predictions based on other features.
Utilize advanced imputation techniques: Advanced methods like multiple imputation, k-nearest neighbors, or matrix completion can be employed to estimate missing values.

7. Describe the various methods for dealing with missing data values in depth.

Listwise deletion: It involves removing instances with missing values, either the entire row or column. It can lead to data loss and bias if missingness is not random.

Mean/median imputation: Missing values are replaced with the mean or median of the available data for the respective attribute. This method assumes no relationship between the missing values and other variables.

Mode imputation: Missing categorical values are replaced with the mode (most frequent value) of the available data for that attribute.

Regression imputation: Missing values are estimated using regression models based on the relationship with other variables.

Multiple imputation: It involves creating multiple imputed datasets by filling missing values using statistical models. These datasets are then analyzed using the chosen machine learning algorithm, and results are combined using specific rules.

9. Explain dimensionality reduction and feature selection in a few words.

Dimensionality reduction: It is the process of reducing the number of features (dimensions) in a dataset while retaining most of the important information. It helps to overcome the curse of dimensionality, reduces computational complexity, and may improve model performance.

Feature selection: Feature selection involves identifying and selecting the most relevant features from the original dataset. It aims to choose a subset of features that are most informative for the predictive model, thereby reducing overfitting, improving model interpretability, and reducing computational resources required.

i. What is the IQR? What criteria are used to assess it?

IQR (Interquartile Range) is a measure of statistical dispersion, representing the range between the first quartile (Q1) and the third quartile (Q3) in a dataset. It provides information about the spread and variability of the middle 50% of the data. IQR is calculated as IQR = Q3 - Q1.
Criteria used to assess IQR:

Outliers: Values below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR are considered outliers.
Skewness: IQR can help assess the symmetry or skewness of a distribution. If the IQR is symmetric around the median, the distribution is relatively symmetrical.

ii. Describe the various components of a box plot in detail. When will the lower whisker surpass the upper whisker in length? How can box plots be used to identify outliers?

A box plot, also known as a box-and-whisker plot, displays the distribution of a continuous variable through five key components:
Median (Q2): The line inside the box represents the median value.
Box: The box represents the interquartile range (IQR), spanning from the first quartile (Q1) to the third quartile (Q3). It contains the middle 50% of the data.
Whiskers: The whiskers extend from the box to the minimum and maximum non-outlier values within a specified range. They typically extend up to 1.5 times the IQR.
Lower whisker: Represents the lowest non-outlier value within the range of the whiskers.
Upper whisker: Represents the highest non-outlier value within the range of the whiskers.
The lower whisker surpasses the upper whisker in length when the distribution is highly skewed to the left (negative skewness). In such cases, the lower whisker may extend further than the upper whisker due to the presence of outliers on the left side of the distribution.

Box plots can be used to identify outliers by:

Observing data points beyond the whiskers, which are considered potential outliers.
Identifying any data points outside the range of Q1 - 1.5 * IQR or Q3 + 1.5 * IQR, which are typically classified as outliers.

10. Make brief notes on any two of the following:

i. Data collected at regular intervals:
Data collected at regular intervals refers to a type of data where observations or measurements are taken at consistent and predetermined time intervals. This data often occurs in time series analysis and can include variables such as stock prices, temperature recordings, or daily sales data. Regularly spaced data facilitates time-based analysis, trend identification, and forecasting.

ii. The gap between the quartiles:
The gap between the quartiles, also known as the interquartile range (IQR), is a measure of dispersion in a dataset. It represents the range between the first quartile (Q1) and the third quartile (Q3). The IQR contains the middle 50% of the data and is used to identify outliers and assess the spread of the distribution. A larger IQR indicates a wider spread or higher variability in the dataset.

iii. Use a cross-tab:
A cross-tabulation (cross-tab) is a tabular summary of data that displays the relationship between two or more categorical variables. It shows the frequency or count of occurrences for each combination of categories. Cross-tabs are useful for exploring associations, dependencies, or patterns between variables. They can help identify relationships and provide insights into the data structure and interactions between categorical variables.

11. Make a comparison between:

i. Data with nominal and ordinal values:

Nominal data: Nominal data represents categorical variables without any inherent order or ranking. Examples include colors, gender categories, or car types. Nominal data can only be classified into different categories, and mathematical operations like sorting or averaging are not meaningful.

Ordinal data: Ordinal data represents categorical variables with an inherent order or ranking. Examples include education levels (e.g., high school, bachelor's, master's) or satisfaction ratings (e.g., very dissatisfied, neutral, very satisfied). Ordinal data can be ranked or ordered, and mathematical operations like median or percentiles are meaningful.

ii. Histogram and box plot:

Histogram: A histogram is a graphical representation that displays the distribution of continuous or discrete data. It consists of a series of bins or intervals along the x-axis and the frequency or count of occurrences on the y-axis. Histograms provide insights into the shape, central tendency, and spread of the data distribution.

Box plot: A box plot, or box-and-whisker plot, summarizes the distribution of continuous data through key statistical measures. It includes a box representing the interquartile range (IQR), a line representing the median, and whiskers extending to the minimum and maximum non-outlier values. Box plots provide information about central tendency, spread, skewness, and outliers in a compact visual format.

iii. The average and median:

Average (Mean): The average, or mean, is a measure of central tendency that calculates the sum of all values divided by the total number of observations. It is sensitive to extreme values and outliers and can be affected by skewed distributions.

Median: The median is another measure of central tendency that represents the middle value in a dataset when arranged in ascending or descending order. It is less sensitive to extreme values and outliers compared to the average. The median is suitable for skewed distributions and provides a robust estimate of the center.

While both the average and median provide measures of central tendency, the choice between them depends on the distribution characteristics and the presence of outliers. The median is preferred when dealing with skewed data or when outliers may significantly influence the average.
