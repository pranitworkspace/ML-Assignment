1. What are the key tasks involved in getting ready to work with machine learning modeling?

Data collection: Gathering the data which can be required for the model.
Data preprocessing: Cleaning the data, handling missing values, dealing with outliers, and transforming the data into a suitable format for modeling.
Feature engineering: Selecting and creating the most relevant features from the available data that can contribute to the model's performance.
Data splitting: Dividing the data into training, validation, and testing sets to assess the model's performance.
Model selection: Choosing the appropriate machine learning algorithm or model based on the problem type and data characteristics.
Model training: Fitting the selected model on the training data to learn patterns and relationships.
Model evaluation: Assessing the performance of the trained model using appropriate evaluation metrics and techniques.
Model optimization: Tuning the model's hyperparameters to improve its performance and generalization.
Deployment: Implementing the trained model into a production environment for making predictions on new, unseen data.

2. What are the different forms of data used in machine learning? Give a specific example for each of them.

Numerical data: Data represented by numerical values, such as height, temperature, or stock prices. Example: Age of individuals.
Categorical data: Data represented by categories or labels, such as gender, color, or product types. Example: Gender of individuals.
Text data: Data represented as text or sequences of characters, such as reviews, tweets, or articles. Example: Customer reviews for a product.
Image data: Data represented as images or visual content, such as photographs, medical scans, or satellite images. Example: Images of handwritten digits for digit recognition.
Time series data: Data represented in a chronological order, such as stock prices over time, sensor readings, or weather data. Example: Daily closing prices of a stock.

3. Distinguish:

i. Numeric vs. categorical attributes:

Numeric attributes are represented by numerical values and support mathematical operations like addition, subtraction, etc. Example: Age, height.
Categorical attributes are represented by categories or labels and don't support mathematical operations. Example: Gender, color.

ii. Feature selection vs. dimensionality reduction:

Feature selection involves selecting the most relevant subset of features from the original feature set to improve model performance and reduce complexity.
Dimensionality reduction aims to transform the original high-dimensional data into a lower-dimensional space while preserving important information. It reduces the number of features but retains the essence of the data.

4. Make quick notes on any two of the following:

i. The histogram:

A histogram is a graphical representation of the distribution of a dataset.
It consists of bars where the height represents the frequency or count of data points falling into each bin or range of values.
It helps visualize the shape, central tendency, spread, and presence of outliers in the data distribution.
Common histogram shapes include bell-shaped (normal distribution), skewed (positively or negatively), and uniform.
Bins in a histogram are predefined intervals or ranges that divide the data and determine the width of the bars.

ii. Use a scatter plot:

A scatter plot is a two-dimensional plot that displays the values of two variables as points on a Cartesian coordinate system.
It visualizes the relationship between two continuous variables and helps identify patterns, trends, clusters, or outliers.
Each point represents an observation with its x and y values, allowing the examination of their joint distribution.
Scatter plots can reveal the presence of linear or non-linear relationships between variables and provide insights into correlation.
Outliers can be detected as data points that significantly deviate from the overall pattern or lie far away from other points.

5. Why is it necessary to investigate data? Is there a discrepancy in how qualitative and quantitative data are explored?

Investigating data is essential to gain a comprehensive understanding of the dataset and its characteristics.
It helps identify patterns, relationships, distributions, anomalies, and potential issues that can influence the modeling process and interpretation of results.
Exploring data allows for data preprocessing steps, feature engineering, and appropriate modeling techniques to be applied.
There can be differences in how qualitative and quantitative data are explored:
Qualitative data exploration often involves techniques like textual analysis, summarization, and visualization to understand patterns and themes.
Quantitative data exploration focuses on statistical analysis, visualizations, and modeling techniques to examine numerical relationships, distributions, and trends.

6. What are the various histogram shapes? What exactly are 'bins'?

Histogram shapes:

Bell-shaped (normal distribution): A symmetric distribution with a peak at the center.
Skewed distribution: Can be positively skewed (tail extends to the right) or negatively skewed (tail extends to the left).
Uniform distribution: All values have equal probability, resulting in a flat histogram with no apparent peaks or valleys.
Bimodal distribution: Two distinct peaks in the histogram, indicating the presence of two separate groups or modes in the data.

Bins in a histogram:

Bins are intervals or ranges that divide the data on the x-axis of a histogram.
They determine the width of the bars and group data points falling within each range together.
The number of bins affects the granularity of the histogram and can impact the visualization and interpretation of the data distribution.
Selecting an appropriate number of bins is important to avoid oversimplification or excessive detail in the histogram representation.

7. How do we deal with data outliers?

Data outliers are extreme values that deviate significantly from the overall pattern of the dataset.
Strategies for dealing with outliers include:
Outlier removal: Removing the outliers from the dataset if they are determined to be errors or irrelevant to the analysis.
Winsorization: Capping the extreme values at a certain percentile to minimize their impact without completely removing them.
Transformation: Applying mathematical transformations (e.g., logarithmic, square root) to make the distribution more symmetric and reduce the influence of outliers.
Binning: Grouping outliers into a separate category or bin to treat them as a distinct group during analysis.
Robust models: Using models that are less sensitive to outliers, such as robust regression or decision trees.
Imputation: Replacing outliers with more reasonable values based on statistical techniques like mean, median, or regression imputation.

8. What are the various measures of central inclination? Why does the mean vary too much from the median in certain data sets?

Measures of central inclination (central tendency):

Mean: The arithmetic average of a set of values, obtained by summing all values and dividing by the total count. It is influenced by extreme values.
Median: The middle value in a sorted dataset, separating the higher and lower halves. It is less affected by extreme values.
Mode: The value(s) that appear most frequently in the dataset. It can be useful for categorical or discrete data.
The mean can vary significantly from the median in certain data sets due to the presence of outliers or skewed distributions:

Outliers: Extreme values can disproportionately affect the mean since it takes into account the magnitude of all values. The median is less affected because it only considers the middle value(s).
Skewed distributions: Skewness can pull the mean towards the tail of the distribution, particularly in positively or negatively skewed datasets. The median is less influenced by skewness.

9. Describe how a scatter plot can be used to investigate bivariate relationships. Is it possible to find outliers using a scatter plot?

A scatter plot is used to visualize the relationship between two continuous variables in a bivariate analysis.
Each data point is represented as a dot on the plot, with the x-axis representing one variable and the y-axis representing the other.
Scatter plots can reveal patterns, trends, correlations, or lack of relationships between the variables.
Outliers can be identified in a scatter plot as data points that lie far away from the general pattern or exhibit unusual behavior.
Outliers may appear as individual points outside the main cluster or as points with extreme values in either the x or y direction.
By examining the scatter plot, one can assess the presence and impact of outliers on the relationship between the variables.

10. Describe how cross-tabs can be used to figure out how two variables are related.

Cross-tabulation, or crosstab, is a tabular summary of data that shows the relationship between two categorical variables.
It allows for the comparison of the distribution of one variable across different categories of another variable.
By creating a contingency table, the frequencies or counts of each combination of categories are displayed.
Cross-tabs can reveal patterns, associations, or dependencies between the variables.
They enable the identification of how the distribution of one variable changes or varies based on the categories of the other variable.
Statistical tests like chi-square test can be performed on cross-tabs to assess the significance of the relationship between the variables.
